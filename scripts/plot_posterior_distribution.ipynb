```python
import json 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
mu = []
sigma = []
for i in range(1, 5):
    with open(f'UCB_distribution_loop_{i}.jsonl') as f:
        lines = f.readlines()
        for i, line in enumerate(lines):
            if i == 0:
                mu.append(json.loads(line.strip()))
            else:
                sigma.append(json.loads(line.strip()))
#sorted(json.loads(line.strip()), reverse=True)
```

```python
for i in range(1, 5):
    mu_mean = np.mean(mu[i-1])
    sigma_mean = np.mean(sigma[i-1])
    mu_std = np.std(mu[i-1])
    sigma_std = np.std(sigma[i-1])

    print(f"Mean of loop {i} mu values:    ", mu_mean)
    print(f"Mean of loop {i} sigma values: ", sigma_mean)
    print(f"Std. of loop {i} mu values:    ", mu_std)
    print(f"Std. of loop {i} sigma values: ", sigma_std)
    print()
```

```python
plt.clf
fig, axs = plt.subplots(1, 2, figsize=(8, 4), sharey=True, tight_layout=True)


axs[0].hist(mu[0], bins=30, label='loop 1', alpha=0.7)
axs[0].hist(mu[1], bins=30, label='loop 2', alpha=0.7)
axs[0].hist(mu[2], bins=40, label='loop 3', alpha=0.7)
axs[0].hist(mu[3], bins=50, label='loop 4', alpha=0.7)
axs[0].set_title('Posterior mean', fontsize = 14)
axs[0].set_yscale('log')
axs[0].set_xlabel('Posterior mean', fontsize = 12)
axs[0].set_ylabel('Frequency', fontsize = 12)

axs[1].hist(sigma[0], bins=5, label='loop 1', alpha=0.7)
axs[1].hist(sigma[1], bins=15, label='loop 2', alpha=0.7)
axs[1].hist(sigma[2], bins=20, label='loop 3', alpha=0.7)
axs[1].hist(sigma[3], bins=25, label='loop 4', alpha=0.7)
axs[1].set_title('Posterior standard deviation', fontsize = 14)
axs[1].set_xlabel('Posterior SD', fontsize = 12)
handles, labels = axs[0].get_legend_handles_labels()

fig.legend(handles, labels, ncol=len(labels), loc='lower center', fontsize = 12, bbox_to_anchor=(0.5, -0.08))
fig.suptitle("Distribution of posterior mean and standard deviation", fontsize = 16)

fig.show()
```
